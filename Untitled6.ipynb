```python
# Verificar o conteúdo da subpasta 'training_set/training_set'
nested_train_folder = '/content/dataset/training_set/training_set'
nested_train_files = os.listdir(nested_train_folder)
print("Conteúdo da pasta 'training_set/training_set':", nested_train_files)
```

```python
# Caminho das pastas de 'dogs' e 'cats'
dogs_path = os.path.join(nested_train_folder, 'dogs')
cats_path = os.path.join(nested_train_folder, 'cats')

# Listar as imagens nas pastas 'dogs' e 'cats'
dog_images = os.listdir(dogs_path)
cat_images = os.listdir(cats_path)

print("Imagens na classe 'dogs':", dog_images)
print("Imagens na classe 'cats':", cat_images)
```

```python
from PIL import Image

# Caminho de uma imagem de exemplo em cada pasta
dog_image_path = os.path.join(dogs_path, dog_images[0])  # Pegando a primeira imagem de 'dogs'
cat_image_path = os.path.join(cats_path, cat_images[0])  # Pegando a primeira imagem de 'cats'

# Carregar as imagens
dog_image = Image.open(dog_image_path)
cat_image = Image.open(cat_image_path)

# Exibir as imagens
dog_image.show(title="Dog Image")
cat_image.show(title="Cat Image")
```

```python
import tensorflow as tf

# Usando o ImageDataGenerator para carregar as imagens e realizar pré-processamento
datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)  # Normalização dos pixels

# Carregar os dados de treinamento a partir da pasta 'training_set/training_set'
train_data = datagen.flow_from_directory(
    nested_train_folder,  # Caminho da pasta com as subpastas 'dogs' e 'cats'
    target_size=(128, 128),  # Redimensionar as imagens para 128x128
    batch_size=32,  # Quantidade de imagens por batch
    class_mode='binary'  # Como temos duas classes, usamos 'binary' (0 ou 1)
)

# Exibir o primeiro batch de imagens
images, labels = next(train_data)
print("Primeiras imagens:", images.shape)
print("Primeiras labels:", labels)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Construindo a rede neural (CNN)
model = models.Sequential([
    # Primeira camada convolucional
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),

    # Segunda camada convolucional
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Camada convolucional adicional
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Camada de flatten para transformar as matrizes em vetores
    layers.Flatten(),

    # Camada densa para classificação
    layers.Dense(128, activation='relu'),

    # Camada de saída (2 classes: 'dogs' e 'cats')
    layers.Dense(1, activation='sigmoid')  # Usamos sigmoid porque temos 2 classes (binary classification)
])

# Exibir a arquitetura do modelo
model.summary()
```

```python
# Treinar o modelo
history = model.fit(
    train_data,  # Dados de treino
    epochs=10,   # Número de épocas (ajuste conforme necessário)
    validation_data=test_data,  # Dados de validação
)
```

```python
# Usando o ImageDataGenerator para carregar as imagens de teste e realizar pré-processamento
test_data = datagen.flow_from_directory(
    test_folder,  # Caminho para a pasta 'test_set' (que deve conter 'dogs' e 'cats')
    target_size=(128, 128),  # Redimensionar as imagens para 128x128
    batch_size=32,  # Quantidade de imagens por batch
    class_mode='binary'  # Como temos duas classes, usamos 'binary' (0 ou 1)
)
```

```python
# Caminho correto para o diretório de treinamento (com subpastas 'dogs' e 'cats')
train_folder = '/content/dataset/training_set/training_set'

# Usar o ImageDataGenerator para carregar as imagens de treinamento
train_data = datagen.flow_from_directory(
    train_folder,  # Caminho correto com subpastas de classes
    target_size=(128, 128),  # Redimensionar as imagens para 128x128
    batch_size=32,  # Quantidade de imagens por batch
    class_mode='binary'  # Como temos 2 classes ('dogs' e 'cats'), usamos 'binary'
)

print("Classes:", train_data.class_indices)
```

```python
# Treinamento do modelo
history = model.fit(
    train_data,  # Dados de treino
    epochs=10,   # Número de épocas (ajuste conforme necessário)
    validation_data=test_data,  # Dados de validação
)
```

```python
# Avaliar o modelo no conjunto de teste
test_loss, test_acc = model.evaluate(test_data)
print(f'Acurácia no conjunto de teste: {test_acc * 100:.2f}%')
```

```python
import matplotlib.pyplot as plt

# Plotando a acurácia e a perda durante o treinamento
plt.figure(figsize=(12, 4))

# Acurácia
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Acurácia de Treinamento')
plt.plot(history.history['val_accuracy'], label='Acurácia de Validação')
plt.title('Acurácia durante o treinamento')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()

# Perda
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perda de Treinamento')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda durante o treinamento')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()

plt.show()
```
