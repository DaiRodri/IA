```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem colorida (substitua 'sua_imagem.jpg' pelo caminho correto da sua imagem)
imagem_colorida = cv2.imread('/sua_imagem.jpg')

# Verificar se a imagem foi carregada corretamente
if imagem_colorida is None:
    print("Erro ao carregar a imagem.")
else:
    # OpenCV carrega a imagem em formato BGR, mas vamos converter para RGB para visualizar corretamente
    imagem_colorida_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

    # Exibir a imagem original
    plt.imshow(imagem_colorida_rgb)
    plt.title('Imagem Original')
    plt.axis('off')
    plt.show()

```

```python
# Converter a imagem para escala de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# Exibir a imagem em escala de cinza
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Escala de Cinza')
plt.axis('off')
plt.show()
```

```python
# Aplicar limiarizao para converter para preto e branco
# O valor 127  o limiar, voc pode ajust-lo para ver o efeito
_, imagem_pb = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# Exibir a imagem em preto e branco
plt.imshow(imagem_pb, cmap='gray')
plt.title('Imagem Preto e Branco')
plt.axis('off')
plt.show()
```

```python
# Salvar as imagens
cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_pb.jpg', imagem_pb)
```
